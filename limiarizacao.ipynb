# Implementação de limiarização de imagens em escala de cinza

```python
def apply_threshold(img: Image, threshold: int, return_array: bool = False) -> Image:

    img_array = np.array(img)

    is_lower = img_array < threshold
    is_greater = img_array >= threshold

    img_array[is_greater] = 0
    img_array[is_lower] = 255

    return img_array if return_array else Image.fromarray(np.uint8(img_array), mode="L")
```
